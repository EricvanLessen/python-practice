```python
# pandas
import pandas as pd
raw_data = pd.read_csv("./flights.csv")
raw_data.head()
```

```python
# use the data as a numpy array
data = raw_data.to_numpy()
```

```python
print('data.shape: {}'.format(data.shape))
print('data.ndim: {}'.format(data.ndim))
sched_dep = data[:,6]
print(sched_dep)
print(type(sched_dep))
```

```python
sched_dep = data[3:8, 6:9]
print(sched_dep)
```

```python
sched_dep = data[:,6]
print('Sum: {}'.format(sched_dep.sum()))
print('Min: {}'.format(sched_dep.min()))
print('Max: {}'.format(sched_dep.max()))
print('Ave: {}'.format(sched_dep.mean()))
```

```python
sched_dep_hours = data[:,6]/60
print('Sum: {}'.format(sched_dep_hours.sum()))
print('Min: {}'.format(sched_dep_hours.min()))
print('Max: {}'.format(sched_dep_hours.max()))
print('Ave: {}'.format(sched_dep_hours.mean()))
print('Standard deviation: {}'.format(sched_dep_hours.std()))
print('Variance: {}'.format(sched_dep_hours.var()))
```

```python
print(sched_dep_hours[sched_dep_hours > 1])
```

3. Visualization

```python
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 20)
y = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y)
plt.show()
```

```python
plt.title('Sin from 0 to 2*pi')
plt.xlabel('input')
plt.ylabel('output')
plt.plot(x,y, label='sine')
plt.plot(x,y2, label='cosine')
plt.legend(loc="lower center")
plt.show()
```

```python
plt.title('Sin from 0 to 2*pi')
plt.xlabel('input')
plt.ylabel('output')
plt.subplot(2,1,1)
plt.plot(x,y, label='sine')
plt.subplot(2,1,2)
plt.plot(x,y2, label='cosine')
plt.legend()
plt.show()
```

```python
plt.title('Sin from 0 to 2*pi')
plt.xlabel('input')
plt.ylabel('output')
plt.plot(x,y, 'ro', label='sine')
plt.plot(x,y2, marker='^', color='g', label='cosine', linewidth=0)
plt.legend(loc="lower center")
plt.show()
```

4. Visualization with Seaborn

```python
import seaborn as sns
print(sns)
```

```python
df.head()
```

```python
sns.relplot(x='ARR_DELAY', y='AIRLINE', data=df)
```

```python
sns.set()
sns.relplot(x='ARR_DELAY', y='AIRLINE', data=df)
```

```python
sns.relplot(x='ARR_DELAY', y='AIRLINE', hue='SCHEDULE_FLIGHT_TIME', data=df)
```

```python
airports_df = df[(df['ORG_AIR']=='LAX')|(df['ORG_AIR']=='DEN')|(df['ORG_AIR']=='JFK')]
sns.catplot(x='ORG_AIR', y='SCHED_DEP', data=airports_df)
```

```python
sns.catplot(x='ORG_AIR', y='SCHED_DEP', kind='violin', data=airports_df)
```

```python
hour_delay = df[df['ARR_DELAY']>60]
sns.jointplot(x='SCHED_DEP', y='ARR_DELAY', data=hour_delay)
```

```python
hour_delay = df[df['ARR_DELAY']>60]
sns.jointplot(x='SCHED_DEP', y='ARR_DELAY', data=hour_delay, kind='kde')
```

```python
g = sns.FacetGrid(airports_df, col='ORG_AIR')
g.map(plt.scatter, 'SCHED_DEP', 'ARR_DELAY')
```

```python
origins_and_airlines = df[(
    (df['ORG_AIR']=='LAX')
    |(df['ORG_AIR']=='DEN')
    |(df['ORG_AIR']=='JFK')
    )& (
    (df['AIRLINE']=='DL')
    |(df['AIRLINE']=='AA')
    |(df['AIRLINE']=='UA')
    )]
g = sns.FacetGrid(origins_and_airlines, col='ORG_AIR', row='AIRLINE')
g.map(plt.scatter, 'SCHED_DEP', 'ARR_DELAY')
```

```python
sns.pairplot(origins_and_airlines)
```

```python
sns.pairplot(origins_and_airlines, vars=['AIRLINE', 'ARR_DELAY', 'ORG_AIR'])
```
